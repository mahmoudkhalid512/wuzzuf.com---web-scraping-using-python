```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

```python
url = "https://wuzzuf.net/search/jobs/?q=Human+Resources&a=navbl"
```

```python
client = urlopen(url)
```

```python
html = client.read()
```

```python
html
```

```python
client.close()
```

```python
soup = bs (html, "html.parser")
```

```python
soup
```

```python
container = soup.find_all("div",{"class": "css-1gatmva e1v1l3u10"})
```

```python
len(container)
```

```python
bs.prettify(container[0])
```

```python
job = container[0].findAll("h2",{"class": "css-m604qf"})
job[0].text
```

```python
company = container[0].findAll("a", {"class": "css-17s97q8"})
company[0].text
```

```python
type = container[0].findAll("span", {"class": "css-1ve4b75"})
type[0].text
```

```python
f = open("wuzzuf.csv", "w")
header = "Job Title, Company Name, Job Type +\n"
f.write(header)
```

```python
for track in container:
    job = track.findAll("h2",{"class": "css-m604qf"})
    job_title = job[0].text.strip()
    company = track.findAll("a", {"class": "css-17s97q8"})
    company_name = company[0].text
    type = track.findAll("span", {"class": "css-1ve4b75"})
    job_type = type[0].text
    f.write(job_title + ", " + company_name + ", " + job_type + "\n")
f.close()
```
